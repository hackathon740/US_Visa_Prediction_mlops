```python
import pandas as pd
from evidently import Report
from evidently import Dataset, DataDefinition
from evidently.descriptors import Sentiment, TextLength, Contains
from evidently.presets import TextEvals
from evidently.presets import DataDriftPreset
```

```python
data_url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston_frame = pd.read_csv(data_url)
```

```python
# Split into reference and current
reference_data = boston_frame.iloc[:200]
current_data = boston_frame.iloc[200:]
```

```python
# Create report
report = Report(
    metrics=[DataDriftPreset()]
)

# Run report
report.run(
    reference_data=reference_data,
    current_data=current_data
)

# Save report

report.save_html("boston_data_drift_report.html")

print("Report generated successfully!")
```
